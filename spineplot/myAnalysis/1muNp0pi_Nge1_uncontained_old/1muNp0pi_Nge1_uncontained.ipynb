```python
import uproot
import pandas as pd

with uproot.open(file) as f:
    # Load both trees
    signal_tree = f['events/nominal/signal']
    selected_tree = f['events/nominal/selected']

    signal_df = signal_tree.arrays(library='pd')
    selected_df = selected_tree.arrays(library='pd')

    print("=== Signal Tree ===")
    print(f"Total signal events: {len(signal_df)}")
    print("Categories:", signal_df['true_category'].value_counts().sort_index())

    print("\n=== Selected Tree ===")
    print(f"Total selected events: {len(selected_df)}")
    print("Categories:", selected_df['true_category'].value_counts().sort_index())

    # Calculate purity
    signal_in_selected = len(selected_df[selected_df['true_category'] == 0.0])
    total_selected = len(selected_df)

    purity = signal_in_selected / total_selected * 100
    print(f"\nPurity = {signal_in_selected}/{total_selected} = {purity:.1f}%")
```
