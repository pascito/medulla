```python
cfg = """
[[this_includes]]
file = 'configurations/common/variables.toml'

[[this_includes]]
file = 'configurations/common/styles.toml'

[samples]
[samples.nominal]
key = "events/nominal/muons"
exposure_type = "pot"
trees = ['events/nominal/muons']

[samples.onbeam]
key = "events/onbeam/muons"
exposure_type = "livetime"
trees = ['events/onbeam/muons']

[samples.offbeam]
key = "events/offbeam/muons"
exposure_type = "livetime"
trees = ['events/offbeam/muons']

[output]
path = '/Users/mueller/Downloads/tmp'

[analysis]
ordinate_sample = 'nominal'
category_branch = 'category'
category_assignment = [[0], [1], [2], [3]]
category_labels = ['Signal', 'Background', 'Cosmic', 'Other']
category_colors = ['C0', 'C1', 'C3', 'C4']
category_types = ['histogram', 'histogram', 'histogram', 'histogram']

[[figure]]
name = 'muon_start_x'
type = 'SimpleFigure'
style = 'selected_icarus'
[[figure.artists]]
type = 'SpineSpectra1D'
variable = 'reco_particle_start_x'
draw_kwargs = {'show_component_number' = true, 'show_component_percentage' = true, 'invert_stack_order' = false}
"""

# Write the string to a file
with open('cfg.toml', 'w') as f:
    f.write(cfg)

file = '../../data/cosmics.root'
analysis = Analysis('cfg.toml', file)

# Read POT/Livetime from the correct paths
import uproot
rf = uproot.open(file)

# Read POT value for nominal (MC)
try:
    pot_tree = rf['events/nominal/POT']
    pot_data = pot_tree.arrays()
    print(f"POT data structure: {pot_data}")

    # Extract the actual POT value
    if 'POT' in pot_data:
        pot_value = pot_data['POT'][0] if len(pot_data['POT']) > 0 else 6.6e20
    else:
        pot_value = 6.6e20  # Fallback
    analysis.override_exposure('nominal', pot_value, 'pot')
    print(f"Set POT for nominal: {pot_value}")
except Exception as e:
    print(f"Could not read POT: {e}")
    analysis.override_exposure('nominal', 6.6e20, 'pot')

# Read Livetime for onbeam data
try:
    livetime_tree = rf['events/nominal/Livetime']  # Use the Livetime tree for nominal
    livetime_data = livetime_tree.arrays()
    print(f"Nominal Livetime structure: {livetime_data}")

    if 'Livetime' in livetime_data:
        livetime_value = livetime_data['Livetime'][0] if len(livetime_data['Livetime']) > 0 else 1000.0
    else:
        livetime_value = 1000.0
    analysis.override_exposure('onbeam', livetime_value, 'livetime')
    print(f"Set livetime for onbeam: {livetime_value}")
except Exception as e:
    print(f"Could not read onbeam livetime: {e}")
    analysis.override_exposure('onbeam', 1000.0, 'livetime')

# For offbeam, try to read from offbeam POT tree
try:
    offbeam_tree = rf['events/offbeam/POT']
    offbeam_data = offbeam_tree.arrays()
    print(f"Offbeam POT structure: {offbeam_data}")

    if 'Livetime' in offbeam_data:
        livetime_value = offbeam_data['Livetime'][0] if len(offbeam_data['Livetime']) > 0 else 1000.0
    else:
        livetime_value = 1000.0
    analysis.override_exposure('offbeam', livetime_value, 'livetime')
    print(f"Set livetime for offbeam: {livetime_value}")
except Exception as e:
    print(f"Could not read offbeam livetime: {e}")
    analysis.override_exposure('offbeam', 1000.0, 'livetime')

analysis.run(close_figs=False)
```
